##### Варіант 2, IRF340 – польовий транзистор з ізольованим затвором і індукованим каналом p типу.

Для дослідження характеристик передачі польового транзистора  **IRF340** зберемо схему, що наведена малюнку 4.1. Залежно від типу транзистора полярність джерел VS1 та VS2 може бути іншою. Як об'єкт управління виберемо джерело живлення VS2 (через пункти меню "Аналіз – Об'єкт управління"). У меню "Контроль об'єкта виділення"  встановлюємо наступні параметри: "Початкове значення" дорівнює "0", "Кінцеве значення" - 30, "Кількість випадків" - 4.

<center><img src="images/Scheme_1.png" width="400"/><h4>Рисунок 4.1 - Схема для дослідження польового транзистора</h4></center>

Вибираємо пункти меню "Аналіз – Аналіз постійного струму – Перехідні характеристики постійного струму...". Встановлюємо параметр "Початкове значення" рівним "0", параметр "Кінцеве значення" рівним 4 В, а параметр "Кількість точок" рівним 1000. У якості вхід. В результаті отримаємо характеристики передачі польового транзистора (рис. 4.2).

<center><img src="images/Graph_1.png" width="400"/><h4>Рисунок 4.2 - Графік характеристик передачі польового транзистора</h4></center>

Для дослідження вихідних характеристик будемо використовувати ту саму схему, яку використовували в попередньому випадку (рис. 4.1). Як об'єкт управління виберемо джерело живлення VS1. У меню "Контроль об'єкта виділення" встановлюємо "Початкове значення" рівним, 2 В, "Кінцеве значення" – 4 В, а "Кількість випадків" – 6.

Вибираємо пункти меню "Аналіз – Аналіз постійного струму – Перехідні характеристики постійного струму...". Встановлюємо "Початкове значення", що дорівнює "0", "Кінцеве значення" – 30 В, "Кількість точок" 1000. У якості вхідної змінної "Ввод" виберемо джерело VS2. В результаті отримаємо вихідні характеристики транзистора (рис. 4.3). 

<center><img src="images/Graph_2.png" width="400"/><h4>Рисунок 4.3 - Графік сімейства вихідних характеристик</h4></center>

За сімействами статичних характеристик знайдемо крутизну транзистора S, вихідну провідність на пологій ділянці вихідної характеристики та опір каналу у відкритому стані (на початковій ділянці вихідної характеристики). При розрахунку параметрів використовуємо вихідну характеристику для середнього значення вихідних струмів (3,6 В з рис.4.3).

Крутизна транзистора S:
$$
S = \frac{\Delta I_{D}}{\Delta U_{GS}} = \frac{924.78 - 46.16}{4 - 3.6} = 2196\ мА/В = 2.2\ А/В\
$$
Вихідна провідність:
$$
G_{out} = \frac{\Delta I_{D}}{\Delta U_{DS}} = \frac{0}{6 - 4.5} = 0\ См (Сіменс)
$$
Опір каналу у відкритому стані:
$$
R_{DS} = \frac{\Delta U_{DS}}{\Delta I_{D}} = \frac{6 - 4.5}{0.461} = 32.53\ Ом
$$

#### Висновок
У ході лабораторної роботи №4 було досліджено властивості польового транзистора IRF340 (n-канальний MOSFET) за допомогою симуляції в TINA-TI. Отримано сімейства характеристик передачі та вихідних характеристик, а також розраховано ключові параметри: крутизну S ≈ 2,2 А/В, вихідну провідність ≈ 0 См та опір каналу R_DS ≈ 33 Ом при V_GS = 3,6 В.
Отримані значення узгоджуються з теорією та паспортними даними, хоча крутизна виявилася нижчою через малий робочий струм, а опір каналу – вищим через недостатню напругу затвора для повного відкриття. Симуляція наочно продемонструвала залежність параметрів MOSFET від режиму роботи та підтвердила важливість правильного вибору робочої точки. Робота сприяла глибшому розумінню принципу дії польових транзисторів.
